```python
import csv
import numpy as ny
```

```python
data = []
with open(r"C:\Users\rabin\Downloads\MER_T07_02A-2020-02-03 (5).csv", 'r') as csvfile:
    file_reader=csv.reader(csvfile, delimiter=',')
    for row in file_reader:
        data.append(row)
data=ny.array(data)
data
```

```python
#  Explore the important attributes like dimension,shape, data type etc, of the array formed.
print(f"The shape of the array is {ny.shape(data)}")
print(f"The dimension of the array is {ny.ndim(data)}")
print(f"The data type of the array is {data.dtype}")
```
